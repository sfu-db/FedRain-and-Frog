```python
import numpy as np
from matplotlib import pyplot as plt

time_min = np.array([7.643666667,37.291,241.8326667,469.4193333,698.7466667,1139.131667])
data_size = np.array([400/1000, 4000/1000, 40000/1000, 80000/1000, 120000/1000, 200000/1000])

plt.plot(data_size, time_min, 'b*-', markersize=5, linewidth=0.8)
plt.xlabel("n (K)", {'fontsize': 13, 'fontweight' : 'bold'})
plt.ylabel("System Running Time (min)", {'fontsize': 13, 'fontweight' : 'bold'})
plt.show()
```
